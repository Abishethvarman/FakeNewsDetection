```python
df_pol['category'] = 'Politics'
df_ent['category'] = 'Entertainment'
df_sport['category'] = 'Sports'

print(df_pol.head())
print(df_ent.head())
print(df_sport.head())
```

```python
df_pol
```

```python
df_pol['corpus'] = df_pol['title'] + ' ' + df_pol['text']
print(df_pol[['title', 'text', 'corpus']].head())
```

```python
df_pol
```

```python
df_pol_fake = df_pol[df_pol['isfake'] == 0]
print(df_pol_fake.head())
```

```python
df_pol_fake.shape[0]
```

```python
num_rows_per_category = 20000

sampled_df_pol_fake = df_pol_fake.sample(n=num_rows_per_category, replace=True, random_state=42)
sampled_df_ent = df_ent.sample(n=num_rows_per_category, replace=True, random_state=42)
sampled_df_sport = df_sport.sample(n=num_rows_per_category, replace=True, random_state=42)

df_fake_classifier = pd.concat([sampled_df_pol_fake, sampled_df_ent, sampled_df_sport], ignore_index=True)

df_fake_classifier = df_fake_classifier.sample(frac=1, random_state=42).reset_index(drop=True)

print("Total rows in df_fake_classifier:", df_fake_classifier.shape[0])
```

```python
df_fake_classifier
```

```python
df_fake_classifier = df_fake_classifier[['corpus', 'category']]
print(df_fake_classifier.head())
```
