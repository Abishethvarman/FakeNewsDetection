```python
df_reliable = pd.read_csv("Reliable.csv")
df_fake = pd.read_csv("Fake.csv")

df_reliable['isfake'] = 1
df_fake['isfake'] = 0

df1 = pd.concat([df_reliable, df_fake]).reset_index(drop = True)

df1.drop(columns = ['date'], inplace = True)
```

```python
combined_df = pd.concat([df1, df_welfake[['title', 'text', 'isfake']]], axis=0)
```
